```python
# import the libraries
import pandas as pd
from datetime import datetime
import re
import random
```

```python
# Create a panda table of one column date containing 100 dates starting the days this code has been running
datelist=pd.DataFrame({"date":pd.date_range(datetime.today(), periods=100)})

```

```python
# re-formatting the date in the form of mm/dd/yyyy
for i in range(len(datelist)):
    
    datelist["date"][i]=str(datelist["date"][i].month)+"/"+str(datelist["date"][i].day)+"/"+str(datelist["date"][i])
    d=re.compile(r"\d+"+"[/]"+r"\d+"+"[/]"+r"\d+")
    match = d.search(str(datelist["date"][i]))
    datelist["date"][i]=match.group()
datelist
```

```python
# a function to generate the the columns of prices in a random way
def ranging(start,stop):
    randomlist = []
    for i in range(0,100):
        n = random.randint(start,stop)*.01
        randomlist.append(n)
    return randomlist
```

```python
# creating and inserting the new columns to the original table
titles = ["Open Price","High Price","Low Price","Close Price","Adj Close Price"]
for counter in range(len(titles)-1):
    datelist.insert(counter+1,titles[counter],ranging(10000,20000),True)
datelist
```

```python
# Inserting the Adj. Close Price
datelist.insert(5,titles[4],ranging(103565,784651),True)
```

```python
# Showing the new form of the table
datelist
```

```python
# creating a new file and copying in it the table
datelist.to_csv("FB21.csv",index=False)
```
